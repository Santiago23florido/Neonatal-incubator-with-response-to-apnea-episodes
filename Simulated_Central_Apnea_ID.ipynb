```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, BatchNormalization, LSTM
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Configuración de semilla para reproducibilidad
np.random.seed(42)
tf.random.set_seed(42)

# Parámetros
FS_NUEVO = 10        # Frecuencia de muestreo de las señales generadas (10 Hz)
MIN_DURACION_APNEA = 4  # Duración mínima de un episodio de apnea (10 segundos)
MAX_DURACION_APNEA = 30  # Duración máxima estimada de un episodio de apnea (60 segundos)
MAX_SEQ_LENGTH = MAX_DURACION_APNEA * FS_NUEVO  # Longitud máxima de secuencia para padding

def cargar_dataset(ruta_dataset="dataset_apnea"):
    """Carga el dataset completo o los archivos individuales."""
    if os.path.exists(os.path.join(ruta_dataset, "dataset_completo.csv")):
        print("Cargando dataset completo...")
        df = pd.read_csv(os.path.join(ruta_dataset, "dataset_completo.csv"))
        return df
    else:
        print("Cargando muestras individuales...")
        dfs = []
        for archivo in os.listdir(ruta_dataset):
            if archivo.startswith("muestra_") and archivo.endswith(".csv"):
                df = pd.read_csv(os.path.join(ruta_dataset, archivo))
                muestra_id = int(archivo.split("_")[1].split(".")[0])
                df["muestra_id"] = muestra_id
                dfs.append(df)
        return pd.concat(dfs)

def identificar_episodios(etiquetas):
    """
    Identifica episodios continuos de apnea y no-apnea en las etiquetas
    Devuelve: lista de tuplas (inicio, fin, tipo_episodio)
    donde tipo_episodio es 1 para apnea y 0 para no-apnea
    """
    episodios = []
    if len(etiquetas) == 0:
        return episodios
    
    # Inicializar con el primer valor
    inicio_actual = 0
    tipo_actual = etiquetas[0]
    
    # Recorrer todas las etiquetas para identificar cambios
    for i in range(1, len(etiquetas)):
        if etiquetas[i] != tipo_actual:
            # Guardar el episodio que termina
            episodios.append((inicio_actual, i-1, tipo_actual))
            # Iniciar nuevo episodio
            inicio_actual = i
            tipo_actual = etiquetas[i]
    
    # Añadir el último episodio
    episodios.append((inicio_actual, len(etiquetas)-1, tipo_actual))
    
    return episodios

def crear_ventanas_adaptativas(df, fs=FS_NUEVO, min_duracion_apnea=MIN_DURACION_APNEA):
    """
    Crea ventanas de tiempo adaptativas basadas en la duración real de cada episodio de apnea.
    Para episodios no-apnea, se divide en segmentos de tamaño comparable a los de apnea.
    """
    X = []  # Señales
    y = []  # Etiquetas (1 si hay apnea, 0 si no)
    duraciones = []  # Duración de cada ventana
    
    # Umbral mínimo de puntos para considerar un episodio válido
    min_puntos = min_duracion_apnea * fs
    
    # Procesar cada muestra por separado
    for muestra_id in df["muestra_id"].unique():
        df_muestra = df[df["muestra_id"] == muestra_id]
        senal = df_muestra["senal"].values
        etiquetas = df_muestra["apnea"].values
        
        # Identificar episodios continuos de apnea y no-apnea
        episodios = identificar_episodios(etiquetas)
        
        for inicio, fin, tipo in episodios:
            duracion = fin - inicio + 1
            
            # Si el episodio es muy corto (< min_duracion_apnea), lo saltamos
            if duracion < min_puntos:
                continue
            
            # Para episodios largos de no-apnea, los dividimos en ventanas de tamaño comparable
            if tipo == 0 and duracion > 3 * min_puntos:
                # Dividir en segmentos de duración similar a episodios de apnea típicas
                segmento_tamano = 2 * min_puntos  # 20 segundos típicos
                
                for seg_inicio in range(inicio, fin, segmento_tamano):
                    seg_fin = min(seg_inicio + segmento_tamano - 1, fin)
                    
                    # Si el segmento final es muy pequeño, lo ignoramos
                    if (seg_fin - seg_inicio + 1) < min_puntos:
                        continue
                    
                    ventana = senal[seg_inicio:seg_fin+1]
                    # Normalización a nivel de ventana
                    ventana = (ventana - np.mean(ventana)) / (np.std(ventana) + 1e-8)
                    
                    X.append(ventana)
                    y.append(0)  # No-apnea
                    duraciones.append(len(ventana))
            else:
                # Para episodios de apnea o episodios cortos de no-apnea, usamos la ventana completa
                ventana = senal[inicio:fin+1]
                # Normalización a nivel de ventana
                ventana = (ventana - np.mean(ventana)) / (np.std(ventana) + 1e-8)
                
                X.append(ventana)
                y.append(tipo)
                duraciones.append(len(ventana))
    
    # Convertir a arrays de numpy
    y = np.array(y)
    duraciones = np.array(duraciones)
    
    print(f"Ventanas creadas: {len(X)}")
    print(f"Distribución: {np.sum(y)} apneas, {len(y) - np.sum(y)} no-apneas")
    print(f"Duración promedio: {np.mean(duraciones)/fs:.2f} segundos")
    print(f"Duración mínima: {np.min(duraciones)/fs:.2f} segundos")
    print(f"Duración máxima: {np.max(duraciones)/fs:.2f} segundos")
    
    return X, y, duraciones

def preparar_datos_para_modelo(X, y, max_seq_length=MAX_SEQ_LENGTH):
    """
    Prepara los datos para el modelo, aplicando padding a las secuencias
    para que todas tengan la misma longitud.
    """
    # Aplicar padding a las secuencias
    X_padded = pad_sequences(
        X, 
        maxlen=max_seq_length,
        dtype='float32',
        padding='post',
        truncating='post',
        value=0.0
    )
    
    # Reshape para CNN: [muestras, tiempo, canales]
    X_padded = X_padded.reshape(X_padded.shape[0], X_padded.shape[1], 1)
    
    return X_padded, np.array(y)

def crear_modelo_cnn_lstm(max_seq_length=MAX_SEQ_LENGTH):
    """Crea un modelo híbrido CNN-LSTM para la detección de apnea con ventanas variables."""
    modelo = Sequential([
        # Capas convolucionales para extraer características
        Conv1D(filters=32, kernel_size=8, activation='relu', input_shape=(max_seq_length, 1),
               padding='same'),
        BatchNormalization(),
        MaxPooling1D(pool_size=4),
        
        Conv1D(filters=64, kernel_size=5, activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling1D(pool_size=4),
        
        # Capa LSTM para manejar la naturaleza secuencial
        LSTM(64, return_sequences=False),
        
        # Capas densas para la clasificación
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')  # Clasificación binaria (0: no apnea, 1: apnea)
    ])
    
    # Compilar el modelo
    modelo.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall(),
                 tf.keras.metrics.AUC()]
    )
    
    return modelo

def entrenar_modelo(X_train, y_train, X_val, y_val, batch_size=32, epochs=50):
    """Entrena el modelo con callbacks para mejorar el rendimiento."""
    modelo = crear_modelo_cnn_lstm()
    
    print(modelo.summary())
    
    # Callbacks
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True
    )
    
    reducir_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=0.00001
    )
    
    checkpoint = ModelCheckpoint(
        'mejor_modelo_apnea_adaptativo.h5',
        monitor='val_auc',
        mode='max',
        save_best_only=True,
        verbose=1
    )
    
    # Entrenamiento con callbacks
    history = modelo.fit(
        X_train, y_train,
        batch_size=batch_size,
        epochs=epochs,
        validation_data=(X_val, y_val),
        callbacks=[early_stopping, reducir_lr, checkpoint],
        verbose=1
    )
    
    return modelo, history

def evaluar_modelo(modelo, X_test, y_test):
    """Evalúa el modelo y genera visualizaciones."""
    # Predicciones
    y_pred_prob = modelo.predict(X_test)
    y_pred = np.round(y_pred_prob)
    
    # Métricas
    print("\nInforme de clasificación:")
    print(classification_report(y_test, y_pred))
    
    conf_matrix = confusion_matrix(y_test, y_pred)
    print("\nMatriz de confusión:")
    print(conf_matrix)
    
    # Calcular curva ROC
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    
    # Visualizaciones
    plt.figure(figsize=(12, 5))
    
    # Gráfica de la matriz de confusión
    plt.subplot(1, 2, 1)
    plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Matriz de Confusión')
    plt.colorbar()
    clases = ['No Apnea', 'Apnea']
    tick_marks = np.arange(len(clases))
    plt.xticks(tick_marks, clases, rotation=45)
    plt.yticks(tick_marks, clases)
    
    # Añadir valores a la matriz
    thresh = conf_matrix.max() / 2
    for i in range(conf_matrix.shape[0]):
        for j in range(conf_matrix.shape[1]):
            plt.text(j, i, format(conf_matrix[i, j], 'd'),
                     horizontalalignment="center",
                     color="white" if conf_matrix[i, j] > thresh else "black")
    
    plt.ylabel('Etiqueta Real')
    plt.xlabel('Etiqueta Predicha')
    
    # Gráfica de la curva ROC
    plt.subplot(1, 2, 2)
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Tasa de Falsos Positivos')
    plt.ylabel('Tasa de Verdaderos Positivos')
    plt.title('Curva ROC')
    plt.legend(loc="lower right")
    
    plt.tight_layout()
    plt.savefig('evaluacion_modelo_apnea_adaptativo.png')
    plt.show()
    
    return y_pred_prob, conf_matrix, roc_auc

def visualizar_predicciones(X_test, y_test, y_pred_prob, duraciones_test, indices=None):
    """Visualiza algunas predicciones para evaluar el rendimiento del modelo."""
    if indices is None:
        # Seleccionar aleatoriamente algunas ventanas para visualizar
        indices = np.random.choice(range(len(X_test)), min(5, len(X_test)), replace=False)
    
    plt.figure(figsize=(15, 10))
    for i, idx in enumerate(indices):
        plt.subplot(len(indices), 1, i+1)
        
        # Determinar la longitud real de la secuencia (excluyendo el padding)
        ventana = X_test[idx].flatten()
        # Identificar donde termina la señal real y empieza el padding (valores 0)
        longitud_real = duraciones_test[idx] if duraciones_test is not None else np.sum(ventana != 0)
        
        # Usar solo la señal real (sin padding)
        ventana_real = ventana[:longitud_real]
        tiempo = np.arange(len(ventana_real)) / FS_NUEVO  # Convertir a segundos
        
        plt.plot(tiempo, ventana_real, 'b-')
        
        # Colorear fondo según predicción
        pred_prob = y_pred_prob[idx][0]
        pred_class = 1 if pred_prob >= 0.5 else 0
        real_class = y_test[idx]
        
        color = 'green' if pred_class == real_class else 'red'
        label = f'Real: {"Apnea" if real_class else "Normal"}, '
        label += f'Pred: {"Apnea" if pred_class else "Normal"} ({pred_prob:.2f})'
        label += f', Duración: {longitud_real/FS_NUEVO:.1f}s'
        
        plt.title(label)
        plt.fill_between(tiempo, min(ventana_real), max(ventana_real), color=color, alpha=0.2)
        
        plt.xlabel('Tiempo (s)')
        plt.ylabel('Señal normalizada')
        plt.grid(True)
    
    plt.tight_layout()
    plt.savefig('visualizacion_predicciones_adaptativas.png')
    plt.show()

def main():
    # 1. Cargar datos
    print("Cargando dataset...")
    df = cargar_dataset()
    
    # 2. Crear ventanas adaptativas
    print("Creando ventanas adaptativas para el entrenamiento...")
    X, y, duraciones = crear_ventanas_adaptativas(df)
    
    # 3. Preparar datos para el modelo (aplicar padding)
    X_padded, y = preparar_datos_para_modelo(X, y)
    print(f"Forma de X después de padding: {X_padded.shape}")
    
    # 4. Dividir en conjuntos de entrenamiento, validación y prueba
    # Dividimos primero los índices para mantener la correspondencia con duraciones
    indices = np.arange(len(X_padded))
    indices_train_val, indices_test, y_train_val, y_test = train_test_split(
        indices, y, test_size=0.2, random_state=42, stratify=y
    )
    indices_train, indices_val, y_train, y_val = train_test_split(
        indices_train_val, y_train_val, test_size=0.2, random_state=42, stratify=y_train_val
    )
    
    # Ahora obtenemos los datos usando los índices
    X_train = X_padded[indices_train]
    X_val = X_padded[indices_val]
    X_test = X_padded[indices_test]
    
    # También mantenemos las duraciones originales para visualización
    duraciones_test = duraciones[indices_test]
    
    print(f"Conjunto de entrenamiento: {X_train.shape}")
    print(f"Conjunto de validación: {X_val.shape}")
    print(f"Conjunto de prueba: {X_test.shape}")
    
    # 5. Entrenar el modelo
    print("\nEntrenando modelo CNN-LSTM...")
    modelo, history = entrenar_modelo(X_train, y_train, X_val, y_val)
    
    # 6. Evaluar el modelo
    print("\nEvaluando modelo...")
    y_pred_prob, conf_matrix, roc_auc = evaluar_modelo(modelo, X_test, y_test)
    
    # 7. Visualizar algunas predicciones
    print("\nVisualizando predicciones...")
    visualizar_predicciones(X_test, y_test, y_pred_prob, duraciones_test)
    
    # 8. Guardar el modelo
    modelo.save('modelo_cnn_lstm_apnea_adaptativo.h5')
    print("\nModelo guardado como 'modelo_cnn_lstm_apnea_adaptativo.h5'")
    
    # 9. Visualizar historia del entrenamiento
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Entrenamiento')
    plt.plot(history.history['val_loss'], label='Validación')
    plt.title('Pérdida durante el entrenamiento')
    plt.xlabel('Época')
    plt.ylabel('Pérdida')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Entrenamiento')
    plt.plot(history.history['val_accuracy'], label='Validación')
    plt.title('Precisión durante el entrenamiento')
    plt.xlabel('Época')
    plt.ylabel('Precisión')
    plt.legend()
    
    plt.tight_layout()
    plt.savefig('historia_entrenamiento_adaptativo.png')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, BatchNormalization, LSTM
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Configuración de semilla para reproducibilidad
np.random.seed(42)
tf.random.set_seed(42)

# Parámetros
FS_NUEVO = 10        # Frecuencia de muestreo de las señales generadas (10 Hz)
MIN_DURACION_APNEA = 4  # Duración mínima de un episodio de apnea (10 segundos)
MAX_DURACION_APNEA = 30  # Duración máxima estimada de un episodio de apnea (60 segundos)
MAX_SEQ_LENGTH = MAX_DURACION_APNEA * FS_NUEVO  # Longitud máxima de secuencia para padding

def cargar_dataset(ruta_dataset="dataset_apnea"):
    """Carga el dataset completo o los archivos individuales."""
    if os.path.exists(os.path.join(ruta_dataset, "dataset_completo.csv")):
        print("Cargando dataset completo...")
        df = pd.read_csv(os.path.join(ruta_dataset, "dataset_completo.csv"))
        return df
    else:
        print("Cargando muestras individuales...")
        dfs = []
        for archivo in os.listdir(ruta_dataset):
            if archivo.startswith("muestra_") and archivo.endswith(".csv"):
                df = pd.read_csv(os.path.join(ruta_dataset, archivo))
                muestra_id = int(archivo.split("_")[1].split(".")[0])
                df["muestra_id"] = muestra_id
                dfs.append(df)
        return pd.concat(dfs)

def identificar_episodios(etiquetas):
    """
    Identifica episodios continuos de apnea y no-apnea en las etiquetas
    Devuelve: lista de tuplas (inicio, fin, tipo_episodio)
    donde tipo_episodio es 1 para apnea y 0 para no-apnea
    """
    episodios = []
    if len(etiquetas) == 0:
        return episodios
    
    # Inicializar con el primer valor
    inicio_actual = 0
    tipo_actual = etiquetas[0]
    
    # Recorrer todas las etiquetas para identificar cambios
    for i in range(1, len(etiquetas)):
        if etiquetas[i] != tipo_actual:
            # Guardar el episodio que termina
            episodios.append((inicio_actual, i-1, tipo_actual))
            # Iniciar nuevo episodio
            inicio_actual = i
            tipo_actual = etiquetas[i]
    
    # Añadir el último episodio
    episodios.append((inicio_actual, len(etiquetas)-1, tipo_actual))
    
    return episodios

def crear_ventanas_adaptativas(df, fs=FS_NUEVO, min_duracion_apnea=MIN_DURACION_APNEA):
    """
    Crea ventanas de tiempo adaptativas basadas en la duración real de cada episodio de apnea.
    Para episodios no-apnea, se divide en segmentos de tamaño comparable a los de apnea.
    """
    X = []  # Señales
    y = []  # Etiquetas (1 si hay apnea, 0 si no)
    duraciones = []  # Duración de cada ventana
    
    # Umbral mínimo de puntos para considerar un episodio válido
    min_puntos = min_duracion_apnea * fs
    
    # Procesar cada muestra por separado
    for muestra_id in df["muestra_id"].unique():
        df_muestra = df[df["muestra_id"] == muestra_id]
        senal = df_muestra["senal"].values
        etiquetas = df_muestra["apnea"].values
        
        # Identificar episodios continuos de apnea y no-apnea
        episodios = identificar_episodios(etiquetas)
        
        for inicio, fin, tipo in episodios:
            duracion = fin - inicio + 1
            
            # Si el episodio es muy corto (< min_duracion_apnea), lo saltamos
            if duracion < min_puntos:
                continue
            
            # Para episodios largos de no-apnea, los dividimos en ventanas de tamaño comparable
            if tipo == 0 and duracion > 3 * min_puntos:
                # Dividir en segmentos de duración similar a episodios de apnea típicas
                segmento_tamano = 2 * min_puntos  # 20 segundos típicos
                
                for seg_inicio in range(inicio, fin, segmento_tamano):
                    seg_fin = min(seg_inicio + segmento_tamano - 1, fin)
                    
                    # Si el segmento final es muy pequeño, lo ignoramos
                    if (seg_fin - seg_inicio + 1) < min_puntos:
                        continue
                    
                    ventana = senal[seg_inicio:seg_fin+1]
                    # Normalización a nivel de ventana
                    ventana = (ventana - np.mean(ventana)) / (np.std(ventana) + 1e-8)
                    
                    X.append(ventana)
                    y.append(0)  # No-apnea
                    duraciones.append(len(ventana))
            else:
                # Para episodios de apnea o episodios cortos de no-apnea, usamos la ventana completa
                ventana = senal[inicio:fin+1]
                # Normalización a nivel de ventana
                ventana = (ventana - np.mean(ventana)) / (np.std(ventana) + 1e-8)
                
                X.append(ventana)
                y.append(tipo)
                duraciones.append(len(ventana))
    
    # Convertir a arrays de numpy
    y = np.array(y)
    duraciones = np.array(duraciones)
    
    print(f"Ventanas creadas: {len(X)}")
    print(f"Distribución: {np.sum(y)} apneas, {len(y) - np.sum(y)} no-apneas")
    print(f"Duración promedio: {np.mean(duraciones)/fs:.2f} segundos")
    print(f"Duración mínima: {np.min(duraciones)/fs:.2f} segundos")
    print(f"Duración máxima: {np.max(duraciones)/fs:.2f} segundos")
    
    return X, y, duraciones

def preparar_datos_para_modelo(X, y, max_seq_length=MAX_SEQ_LENGTH):
    """
    Prepara los datos para el modelo, aplicando padding a las secuencias
    para que todas tengan la misma longitud.
    """
    # Aplicar padding a las secuencias
    X_padded = pad_sequences(
        X, 
        maxlen=max_seq_length,
        dtype='float32',
        padding='post',
        truncating='post',
        value=0.0
    )
    
    # Reshape para CNN: [muestras, tiempo, canales]
    X_padded = X_padded.reshape(X_padded.shape[0], X_padded.shape[1], 1)
    
    return X_padded, np.array(y)

def crear_modelo_cnn_lstm(max_seq_length=MAX_SEQ_LENGTH):
    """Crea un modelo híbrido CNN-LSTM para la detección de apnea con ventanas variables."""
    modelo = Sequential([
        # Capas convolucionales para extraer características
        Conv1D(filters=32, kernel_size=8, activation='relu', input_shape=(max_seq_length, 1),
               padding='same'),
        BatchNormalization(),
        MaxPooling1D(pool_size=4),
        
        Conv1D(filters=64, kernel_size=5, activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling1D(pool_size=4),
        
        # Capa LSTM para manejar la naturaleza secuencial
        LSTM(64, return_sequences=False),
        
        # Capas densas para la clasificación
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')  # Clasificación binaria (0: no apnea, 1: apnea)
    ])
    
    # Compilar el modelo
    modelo.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall(),
                 tf.keras.metrics.AUC()]
    )
    
    return modelo

def entrenar_modelo(X_train, y_train, X_val, y_val, batch_size=32, epochs=50):
    """Entrena el modelo con callbacks para mejorar el rendimiento."""
    modelo = crear_modelo_cnn_lstm()
    
    print(modelo.summary())
    
    # Callbacks
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True
    )
    
    reducir_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=0.00001
    )
    
    checkpoint = ModelCheckpoint(
        'mejor_modelo_apnea_adaptativo.h5',
        monitor='val_auc',
        mode='max',
        save_best_only=True,
        verbose=1
    )
    
    # Entrenamiento con callbacks
    history = modelo.fit(
        X_train, y_train,
        batch_size=batch_size,
        epochs=epochs,
        validation_data=(X_val, y_val),
        callbacks=[early_stopping, reducir_lr, checkpoint],
        verbose=1
    )
    
    return modelo, history

def evaluar_modelo(modelo, X_test, y_test):
    """Evalúa el modelo y genera visualizaciones."""
    # Predicciones
    y_pred_prob = modelo.predict(X_test)
    y_pred = np.round(y_pred_prob)
    
    # Métricas
    print("\nInforme de clasificación:")
    print(classification_report(y_test, y_pred))
    
    conf_matrix = confusion_matrix(y_test, y_pred)
    print("\nMatriz de confusión:")
    print(conf_matrix)
    
    # Calcular curva ROC
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    
    # Visualizaciones
    plt.figure(figsize=(12, 5))
    
    # Gráfica de la matriz de confusión
    plt.subplot(1, 2, 1)
    plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion matrix')
    plt.colorbar()
    clases = ['No Apnea', 'Apnea']
    tick_marks = np.arange(len(clases))
    plt.xticks(tick_marks, clases, rotation=45)
    plt.yticks(tick_marks, clases)
    
    # Añadir valores a la matriz
    thresh = conf_matrix.max() / 2
    for i in range(conf_matrix.shape[0]):
        for j in range(conf_matrix.shape[1]):
            plt.text(j, i, format(conf_matrix[i, j], 'd'),
                     horizontalalignment="center",
                     color="white" if conf_matrix[i, j] > thresh else "black")
    
    plt.ylabel('Real')
    plt.xlabel('Predicted')
    
    # Gráfica de la curva ROC
    plt.subplot(1, 2, 2)
    plt.plot(fpr, tpr, color='k', linestyle='-', lw=2, label=f'ROC curve (area = {roc_auc:.2f})') # Separar color y estilo
    plt.plot([0, 1], [0, 1], color='k', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', color='black')
    plt.ylabel('True Positive Rate', color='black')
    plt.title('ROC Curve', color='black')
    plt.legend(loc="lower right", labelcolor='black')
    
    plt.tight_layout()
    plt.savefig('evaluacion_modelo_apnea_adaptativo.png')
    plt.show()
    
    return y_pred_prob, conf_matrix, roc_auc

def visualizar_predicciones(X_test, y_test, y_pred_prob, duraciones_test, indices=None):
    """Visualiza algunas predicciones para evaluar el rendimiento del modelo."""
    if indices is None:
        # Seleccionar aleatoriamente algunas ventanas para visualizar
        indices = np.random.choice(range(len(X_test)), min(5, len(X_test)), replace=False)
    
    plt.figure(figsize=(15, 10))
    for i, idx in enumerate(indices):
        plt.subplot(len(indices), 1, i+1)
        
        # Determinar la longitud real de la secuencia (excluyendo el padding)
        ventana = X_test[idx].flatten()
        # Identificar donde termina la señal real y empieza el padding (valores 0)
        longitud_real = duraciones_test[idx] if duraciones_test is not None else np.sum(ventana != 0)
        
        # Usar solo la señal real (sin padding)
        ventana_real = ventana[:longitud_real]
        tiempo = np.arange(len(ventana_real)) / FS_NUEVO  # Convertir a segundos
        
        plt.plot(tiempo, ventana_real, 'b-')
        
        # Colorear fondo según predicción
        pred_prob = y_pred_prob[idx][0]
        pred_class = 1 if pred_prob >= 0.5 else 0
        real_class = y_test[idx]
        
        color = 'green' if pred_class == real_class else 'red'
        label = f'Real: {"Apnea" if real_class else "Normal"}, '
        label += f'Pred: {"Apnea" if pred_class else "Normal"} ({pred_prob:.2f})'
        label += f', Duración: {longitud_real/FS_NUEVO:.1f}s'
        
        plt.title(label)
        plt.fill_between(tiempo, min(ventana_real), max(ventana_real), color=color, alpha=0.2)
        
        plt.xlabel('Tiempo (s)')
        plt.ylabel('Señal normalizada')
        plt.grid(True)
    
    plt.tight_layout()
    plt.savefig('visualizacion_predicciones_adaptativas.png')
    plt.show()

def main():
    # 1. Cargar datos
    print("Cargando dataset...")
    df = cargar_dataset()
    
    # 2. Crear ventanas adaptativas
    print("Creando ventanas adaptativas para el entrenamiento...")
    X, y, duraciones = crear_ventanas_adaptativas(df)
    
    # 3. Preparar datos para el modelo (aplicar padding)
    X_padded, y = preparar_datos_para_modelo(X, y)
    print(f"Forma de X después de padding: {X_padded.shape}")
    
    # 4. Dividir en conjuntos de entrenamiento, validación y prueba
    # Dividimos primero los índices para mantener la correspondencia con duraciones
    indices = np.arange(len(X_padded))
    indices_train_val, indices_test, y_train_val, y_test = train_test_split(
        indices, y, test_size=0.2, random_state=42, stratify=y
    )
    indices_train, indices_val, y_train, y_val = train_test_split(
        indices_train_val, y_train_val, test_size=0.2, random_state=42, stratify=y_train_val
    )
    
    # Ahora obtenemos los datos usando los índices
    X_train = X_padded[indices_train]
    X_val = X_padded[indices_val]
    X_test = X_padded[indices_test]
    
    # También mantenemos las duraciones originales para visualización
    duraciones_test = duraciones[indices_test]
    
    print(f"Conjunto de entrenamiento: {X_train.shape}")
    print(f"Conjunto de validación: {X_val.shape}")
    print(f"Conjunto de prueba: {X_test.shape}")
    
    # 5. Entrenar el modelo
    print("\nEntrenando modelo CNN-LSTM...")
    modelo, history = entrenar_modelo(X_train, y_train, X_val, y_val)
    
    # 6. Evaluar el modelo
    print("\nEvaluando modelo...")
    y_pred_prob, conf_matrix, roc_auc = evaluar_modelo(modelo, X_test, y_test)
    
    # 7. Visualizar algunas predicciones
    print("\nVisualizando predicciones...")
    visualizar_predicciones(X_test, y_test, y_pred_prob, duraciones_test)
    
    # 8. Guardar el modelo
    modelo.save('ignorar_cnn_lstm_apnea_adaptativo.h5')
    print("\nModelo guardado como 'modelo_cnn_lstm_apnea_adaptativo.h5'")
    
    # 9. Visualizar historia del entrenamiento
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training', linestyle='-', color='black')
    plt.plot(history.history['val_loss'], label='Validation', linestyle='--', color='black')
    plt.title('Loss during training', color='black')
    plt.xlabel('Epoch', color='black')
    plt.ylabel('Loss', color='black')
    plt.legend(labelcolor='black')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training', linestyle='-', color='black')
    plt.plot(history.history['val_accuracy'], label='Validation', linestyle='--', color='black')
    plt.title('Accuracy during training', color='black')
    plt.xlabel('Epoch', color='black')
    plt.ylabel('Accuracy', color='black')
    plt.legend(labelcolor='black')

if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, BatchNormalization, LSTM
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Configuración de semilla para reproducibilidad
np.random.seed(42)
tf.random.set_seed(42)

# Parámetros
FS_NUEVO = 10        # Frecuencia de muestreo de las señales generadas (10 Hz)
MIN_DURACION_APNEA = 4  # Duración mínima de un episodio de apnea (10 segundos)
MAX_DURACION_APNEA = 8  # Duración máxima estimada de un episodio de apnea (60 segundos)
MAX_SEQ_LENGTH = MAX_DURACION_APNEA * FS_NUEVO  # Longitud máxima de secuencia para padding

def cargar_dataset(ruta_dataset="dataset_apnea"):
    """Carga el dataset completo o los archivos individuales."""
    if os.path.exists(os.path.join(ruta_dataset, "dataset_completo.csv")):
        print("Cargando dataset completo...")
        df = pd.read_csv(os.path.join(ruta_dataset, "dataset_completo.csv"))
        return df
    else:
        print("Cargando muestras individuales...")
        dfs = []
        for archivo in os.listdir(ruta_dataset):
            if archivo.startswith("muestra_") and archivo.endswith(".csv"):
                df = pd.read_csv(os.path.join(ruta_dataset, archivo))
                muestra_id = int(archivo.split("_")[1].split(".")[0])
                df["muestra_id"] = muestra_id
                dfs.append(df)
        return pd.concat(dfs)

def identificar_episodios(etiquetas):
    """
    Identifica episodios continuos de apnea y no-apnea en las etiquetas
    Devuelve: lista de tuplas (inicio, fin, tipo_episodio)
    donde tipo_episodio es 1 para apnea y 0 para no-apnea
    """
    episodios = []
    if len(etiquetas) == 0:
        return episodios
    
    # Inicializar con el primer valor
    inicio_actual = 0
    tipo_actual = etiquetas[0]
    
    # Recorrer todas las etiquetas para identificar cambios
    for i in range(1, len(etiquetas)):
        if etiquetas[i] != tipo_actual:
            # Guardar el episodio que termina
            episodios.append((inicio_actual, i-1, tipo_actual))
            # Iniciar nuevo episodio
            inicio_actual = i
            tipo_actual = etiquetas[i]
    
    # Añadir el último episodio
    episodios.append((inicio_actual, len(etiquetas)-1, tipo_actual))
    
    return episodios

def crear_ventanas_adaptativas(df, fs=FS_NUEVO, duracion_ventana=8):
    
    X = []  # Señales
    y = []  # Etiquetas (1 si hay apnea, 0 si no)
    duraciones = []  # Duración de cada ventana
    puntos_por_ventana = duracion_ventana * fs

    # Procesar cada muestra por separado
    for muestra_id in df["muestra_id"].unique():
        df_muestra = df[df["muestra_id"] == muestra_id]
        senal = df_muestra["senal"].values
        etiquetas = df_muestra["apnea"].values

        # Iterar a través de la señal en ventanas de tamaño fijo
        for i in range(0, len(senal), puntos_por_ventana):
            inicio = i
            fin = i + puntos_por_ventana - 1

            # Asegurarse de que no nos salgamos del final de la señal
            if fin >= len(senal):
                # Si la última ventana es más corta que la deseada, la podemos descartar
                # o rellenar. Para este caso, la descartaremos si es demasiado pequeña.
                if (len(senal) - inicio) < (puntos_por_ventana / 2): # Example: discard if less than half
                    continue
                fin = len(senal) - 1

            ventana_senal = senal[inicio:fin+1]
            ventana_etiquetas = etiquetas[inicio:fin+1]

            # Normalización a nivel de ventana
            # Evitar división por cero si la desviación estándar es muy pequeña
            ventana_senal = (ventana_senal - np.mean(ventana_senal)) / (np.std(ventana_senal) + 1e-8)

            # Etiquetar la ventana: 1 si TODOS los puntos son apnea, 0 si no
            etiqueta_ventana = 1 if np.all(ventana_etiquetas == 1) else 0

            X.append(ventana_senal)
            y.append(etiqueta_ventana)
            duraciones.append(len(ventana_senal))
    
    # Convertir a arrays de numpy
    y = np.array(y)
    duraciones = np.array(duraciones)

    print(f"Ventanas creadas: {len(X)}")
    print(f"Distribución: {np.sum(y)} apneas, {len(y) - np.sum(y)} no-apneas")
    print(f"Duración promedio: {np.mean(duraciones)/fs:.2f} segundos")
    print(f"Duración mínima: {np.min(duraciones)/fs:.2f} segundos")
    print(f"Duración máxima: {np.max(duraciones)/fs:.2f} segundos")

    return X, y, duraciones

def preparar_datos_para_modelo(X, y, max_seq_length=MAX_SEQ_LENGTH):
    """
    Prepara los datos para el modelo, aplicando padding a las secuencias
    para que todas tengan la misma longitud.
    """
    # Aplicar padding a las secuencias
    X_padded = pad_sequences(
        X, 
        maxlen=max_seq_length,
        dtype='float32',
        padding='post',
        truncating='post',
        value=0.0
    )
    
    # Reshape para CNN: [muestras, tiempo, canales]
    X_padded = X_padded.reshape(X_padded.shape[0], X_padded.shape[1], 1)
    
    return X_padded, np.array(y)

def crear_modelo_cnn_lstm(max_seq_length=MAX_SEQ_LENGTH):
    """Crea un modelo híbrido CNN-LSTM para la detección de apnea con ventanas variables."""
    modelo = Sequential([
        # Capas convolucionales para extraer características
        Conv1D(filters=32, kernel_size=8, activation='relu', input_shape=(max_seq_length, 1),
               padding='same'),
        BatchNormalization(),
        MaxPooling1D(pool_size=4),
        
        Conv1D(filters=64, kernel_size=5, activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling1D(pool_size=4),
        
        # Capa LSTM para manejar la naturaleza secuencial
        LSTM(64, return_sequences=False),
        
        # Capas densas para la clasificación
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')  # Clasificación binaria (0: no apnea, 1: apnea)
    ])
    
    # Compilar el modelo
    modelo.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall(),
                 tf.keras.metrics.AUC()]
    )
    
    return modelo

def entrenar_modelo(X_train, y_train, X_val, y_val, batch_size=32, epochs=50):
    """Entrena el modelo con callbacks para mejorar el rendimiento."""
    modelo = crear_modelo_cnn_lstm()
    
    print(modelo.summary())
    
    # Callbacks
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True
    )
    
    reducir_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=0.00001
    )
    
    checkpoint = ModelCheckpoint(
        'mejor_modelo_apnea_adaptativo.h5',
        monitor='val_auc',
        mode='max',
        save_best_only=True,
        verbose=1
    )
    
    # Entrenamiento con callbacks
    history = modelo.fit(
        X_train, y_train,
        batch_size=batch_size,
        epochs=epochs,
        validation_data=(X_val, y_val),
        callbacks=[early_stopping, reducir_lr, checkpoint],
        verbose=1
    )
    
    return modelo, history

def evaluar_modelo(modelo, X_test, y_test):
    """Evalúa el modelo y genera visualizaciones."""
    # Predicciones
    y_pred_prob = modelo.predict(X_test)
    y_pred = np.round(y_pred_prob)
    
    # Métricas
    print("\nInforme de clasificación:")
    print(classification_report(y_test, y_pred))
    
    conf_matrix = confusion_matrix(y_test, y_pred)
    print("\nMatriz de confusión:")
    print(conf_matrix)
    
    # Calcular curva ROC
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    
    # Visualizaciones
    plt.figure(figsize=(12, 5))
    
    # Gráfica de la matriz de confusión
    plt.subplot(1, 2, 1)
    plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion matrix')
    plt.colorbar()
    clases = ['No Apnea', 'Apnea']
    tick_marks = np.arange(len(clases))
    plt.xticks(tick_marks, clases, rotation=45)
    plt.yticks(tick_marks, clases)
    
    # Añadir valores a la matriz
    thresh = conf_matrix.max() / 2
    for i in range(conf_matrix.shape[0]):
        for j in range(conf_matrix.shape[1]):
            plt.text(j, i, format(conf_matrix[i, j], 'd'),
                     horizontalalignment="center",
                     color="white" if conf_matrix[i, j] > thresh else "black")
    
    plt.ylabel('Real')
    plt.xlabel('Predicted')
    
    # Gráfica de la curva ROC
    plt.subplot(1, 2, 2)
    plt.plot(fpr, tpr, color='k', linestyle='-', lw=2, label=f'ROC curve (area = {roc_auc:.2f})') # Separar color y estilo
    plt.plot([0, 1], [0, 1], color='k', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', color='black')
    plt.ylabel('True Positive Rate', color='black')
    plt.title('ROC Curve', color='black')
    plt.legend(loc="lower right", labelcolor='black')
    
    plt.tight_layout()
    plt.savefig('evaluacion_modelo_apnea_adaptativo.png')
    plt.show()
    
    return y_pred_prob, conf_matrix, roc_auc

def visualizar_predicciones(X_test, y_test, y_pred_prob, duraciones_test, indices=None):
    """Visualiza algunas predicciones para evaluar el rendimiento del modelo."""
    if indices is None:
        # Seleccionar aleatoriamente algunas ventanas para visualizar
        indices = np.random.choice(range(len(X_test)), min(5, len(X_test)), replace=False)
    
    plt.figure(figsize=(15, 10))
    for i, idx in enumerate(indices):
        plt.subplot(len(indices), 1, i+1)
        
        # Determinar la longitud real de la secuencia (excluyendo el padding)
        ventana = X_test[idx].flatten()
        # Identificar donde termina la señal real y empieza el padding (valores 0)
        longitud_real = duraciones_test[idx] if duraciones_test is not None else np.sum(ventana != 0)
        
        # Usar solo la señal real (sin padding)
        ventana_real = ventana[:longitud_real]
        tiempo = np.arange(len(ventana_real)) / FS_NUEVO  # Convertir a segundos
        
        plt.plot(tiempo, ventana_real, 'b-')
        
        # Colorear fondo según predicción
        pred_prob = y_pred_prob[idx][0]
        pred_class = 1 if pred_prob >= 0.5 else 0
        real_class = y_test[idx]
        
        color = 'green' if pred_class == real_class else 'red'
        label = f'Real: {"Apnea" if real_class else "Normal"}, '
        label += f'Pred: {"Apnea" if pred_class else "Normal"} ({pred_prob:.2f})'
        label += f', Duración: {longitud_real/FS_NUEVO:.1f}s'
        
        plt.title(label)
        plt.fill_between(tiempo, min(ventana_real), max(ventana_real), color=color, alpha=0.2)
        
        plt.xlabel('Tiempo (s)')
        plt.ylabel('Señal normalizada')
        plt.grid(True)
    
    plt.tight_layout()
    plt.savefig('visualizacion_predicciones_adaptativas.png')
    plt.show()

def main():
    # 1. Cargar datos
    print("Cargando dataset...")
    df = cargar_dataset()
    
    # 2. Crear ventanas adaptativas
    print("Creando ventanas adaptativas para el entrenamiento...")
    X, y, duraciones = crear_ventanas_adaptativas(df)
    
    # 3. Preparar datos para el modelo (aplicar padding)
    X_padded, y = preparar_datos_para_modelo(X, y)
    print(f"Forma de X después de padding: {X_padded.shape}")
    
    # 4. Dividir en conjuntos de entrenamiento, validación y prueba
    # Dividimos primero los índices para mantener la correspondencia con duraciones
    indices = np.arange(len(X_padded))
    indices_train_val, indices_test, y_train_val, y_test = train_test_split(
        indices, y, test_size=0.2, random_state=42, stratify=y
    )
    indices_train, indices_val, y_train, y_val = train_test_split(
        indices_train_val, y_train_val, test_size=0.2, random_state=42, stratify=y_train_val
    )
    
    # Ahora obtenemos los datos usando los índices
    X_train = X_padded[indices_train]
    X_val = X_padded[indices_val]
    X_test = X_padded[indices_test]
    
    # También mantenemos las duraciones originales para visualización
    duraciones_test = duraciones[indices_test]
    
    print(f"Conjunto de entrenamiento: {X_train.shape}")
    print(f"Conjunto de validación: {X_val.shape}")
    print(f"Conjunto de prueba: {X_test.shape}")
    
    # 5. Entrenar el modelo
    print("\nEntrenando modelo CNN-LSTM...")
    modelo, history = entrenar_modelo(X_train, y_train, X_val, y_val)
    
    # 6. Evaluar el modelo
    print("\nEvaluando modelo...")
    y_pred_prob, conf_matrix, roc_auc = evaluar_modelo(modelo, X_test, y_test)
    
    # 7. Visualizar algunas predicciones
    print("\nVisualizando predicciones...")
    visualizar_predicciones(X_test, y_test, y_pred_prob, duraciones_test)
    
    # 8. Guardar el modelo
    modelo.save('ignorar_cnn_lstm_apnea_adaptativo_ventanas_fijas.h5')
    print("\nModelo guardado como 'modelo_cnn_lstm_apnea_adaptativo.h5'")
    
    # 9. Visualizar historia del entrenamiento
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training', linestyle='-', color='black')
    plt.plot(history.history['val_loss'], label='Validation', linestyle='--', color='black')
    plt.title('Loss during training', color='black')
    plt.xlabel('Epoch', color='black')
    plt.ylabel('Loss', color='black')
    plt.legend(labelcolor='black')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training', linestyle='-', color='black')
    plt.plot(history.history['val_accuracy'], label='Validation', linestyle='--', color='black')
    plt.title('Accuracy during training', color='black')
    plt.xlabel('Epoch', color='black')
    plt.ylabel('Accuracy', color='black')
    plt.legend(labelcolor='black')

if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, BatchNormalization, LSTM
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Configuración de semilla para reproducibilidad
np.random.seed(42)
tf.random.set_seed(42)

# Parámetros
FS_NUEVO = 10        # Frecuencia de muestreo de las señales generadas (10 Hz)
MIN_DURACION_APNEA = 2  # Duración mínima de un episodio de apnea (10 segundos)
MAX_DURACION_APNEA = 8  # Duración máxima estimada de un episodio de apnea (60 segundos)
MAX_SEQ_LENGTH = MAX_DURACION_APNEA * FS_NUEVO  # Longitud máxima de secuencia para padding

def cargar_dataset(ruta_dataset="dataset_apnea"):
    """Carga el dataset completo o los archivos individuales."""
    if os.path.exists(os.path.join(ruta_dataset, "dataset_completo.csv")):
        print("Cargando dataset completo...")
        df = pd.read_csv(os.path.join(ruta_dataset, "dataset_completo.csv"))
        return df
    else:
        print("Cargando muestras individuales...")
        dfs = []
        for archivo in os.listdir(ruta_dataset):
            if archivo.startswith("muestra_") and archivo.endswith(".csv"):
                df = pd.read_csv(os.path.join(ruta_dataset, archivo))
                muestra_id = int(archivo.split("_")[1].split(".")[0])
                df["muestra_id"] = muestra_id
                dfs.append(df)
        return pd.concat(dfs)

def identificar_episodios(etiquetas):
    """
    Identifica episodios continuos de apnea y no-apnea en las etiquetas
    Devuelve: lista de tuplas (inicio, fin, tipo_episodio)
    donde tipo_episodio es 1 para apnea y 0 para no-apnea
    """
    episodios = []
    if len(etiquetas) == 0:
        return episodios
    
    # Inicializar con el primer valor
    inicio_actual = 0
    tipo_actual = etiquetas[0]
    
    # Recorrer todas las etiquetas para identificar cambios
    for i in range(1, len(etiquetas)):
        if etiquetas[i] != tipo_actual:
            # Guardar el episodio que termina
            episodios.append((inicio_actual, i-1, tipo_actual))
            # Iniciar nuevo episodio
            inicio_actual = i
            tipo_actual = etiquetas[i]
    
    # Añadir el último episodio
    episodios.append((inicio_actual, len(etiquetas)-1, tipo_actual))
    
    return episodios

def crear_ventanas_adaptativas(df, fs=FS_NUEVO, min_duracion_apnea=MIN_DURACION_APNEA):
    """
    Crea ventanas de tiempo adaptativas basadas en la duración real de cada episodio de apnea.
    Para episodios no-apnea, se divide en segmentos de tamaño comparable a los de apnea.
    """
    X = []  # Señales
    y = []  # Etiquetas (1 si hay apnea, 0 si no)
    duraciones = []  # Duración de cada ventana
    
    # Umbral mínimo de puntos para considerar un episodio válido
    min_puntos = min_duracion_apnea * fs
    
    # Procesar cada muestra por separado
    for muestra_id in df["muestra_id"].unique():
        df_muestra = df[df["muestra_id"] == muestra_id]
        senal = df_muestra["senal"].values
        etiquetas = df_muestra["apnea"].values
        
        # Identificar episodios continuos de apnea y no-apnea
        episodios = identificar_episodios(etiquetas)
        
        for inicio, fin, tipo in episodios:
            duracion = fin - inicio + 1
            
            # Si el episodio es muy corto (< min_duracion_apnea), lo saltamos
            if duracion < min_puntos:
                continue
            
            # Para episodios largos de no-apnea, los dividimos en ventanas de tamaño comparable
            if tipo == 0 and duracion > 3 * min_puntos:
                # Dividir en segmentos de duración similar a episodios de apnea típicas
                segmento_tamano = 2 * min_puntos  # 20 segundos típicos
                
                for seg_inicio in range(inicio, fin, segmento_tamano):
                    seg_fin = min(seg_inicio + segmento_tamano - 1, fin)
                    
                    # Si el segmento final es muy pequeño, lo ignoramos
                    if (seg_fin - seg_inicio + 1) < min_puntos:
                        continue
                    
                    ventana = senal[seg_inicio:seg_fin+1]
                    # Normalización a nivel de ventana
                    ventana = (ventana - np.mean(ventana)) / (np.std(ventana) + 1e-8)
                    
                    X.append(ventana)
                    y.append(0)  # No-apnea
                    duraciones.append(len(ventana))
            else:
                # Para episodios de apnea o episodios cortos de no-apnea, usamos la ventana completa
                ventana = senal[inicio:fin+1]
                # Normalización a nivel de ventana
                ventana = (ventana - np.mean(ventana)) / (np.std(ventana) + 1e-8)
                
                X.append(ventana)
                y.append(tipo)
                duraciones.append(len(ventana))
    
    # Convertir a arrays de numpy
    y = np.array(y)
    duraciones = np.array(duraciones)
    
    print(f"Ventanas creadas: {len(X)}")
    print(f"Distribución: {np.sum(y)} apneas, {len(y) - np.sum(y)} no-apneas")
    print(f"Duración promedio: {np.mean(duraciones)/fs:.2f} segundos")
    print(f"Duración mínima: {np.min(duraciones)/fs:.2f} segundos")
    print(f"Duración máxima: {np.max(duraciones)/fs:.2f} segundos")
    
    return X, y, duraciones

def preparar_datos_para_modelo(X, y, max_seq_length=MAX_SEQ_LENGTH):
    """
    Prepara los datos para el modelo, aplicando padding a las secuencias
    para que todas tengan la misma longitud.
    """
    # Aplicar padding a las secuencias
    X_padded = pad_sequences(
        X, 
        maxlen=max_seq_length,
        dtype='float32',
        padding='post',
        truncating='post',
        value=0.0
    )
    
    # Reshape para CNN: [muestras, tiempo, canales]
    X_padded = X_padded.reshape(X_padded.shape[0], X_padded.shape[1], 1)
    
    return X_padded, np.array(y)

def crear_modelo_cnn_lstm(max_seq_length=MAX_SEQ_LENGTH):
    """Crea un modelo híbrido CNN-LSTM para la detección de apnea con ventanas variables."""
    modelo = Sequential([
        # Capas convolucionales para extraer características
        Conv1D(filters=32, kernel_size=8, activation='relu', input_shape=(max_seq_length, 1),
               padding='same'),
        BatchNormalization(),
        MaxPooling1D(pool_size=4),
        
        Conv1D(filters=64, kernel_size=5, activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling1D(pool_size=4),
        
        # Capa LSTM para manejar la naturaleza secuencial
        LSTM(64, return_sequences=False),
        
        # Capas densas para la clasificación
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')  # Clasificación binaria (0: no apnea, 1: apnea)
    ])
    
    # Compilar el modelo
    modelo.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall(),
                 tf.keras.metrics.AUC()]
    )
    
    return modelo

def entrenar_modelo(X_train, y_train, X_val, y_val, batch_size=32, epochs=50):
    """Entrena el modelo con callbacks para mejorar el rendimiento."""
    modelo = crear_modelo_cnn_lstm()
    
    print(modelo.summary())
    
    # Callbacks
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True
    )
    
    reducir_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=0.00001
    )
    
    checkpoint = ModelCheckpoint(
        'mejor_modelo_apnea_adaptativo.h5',
        monitor='val_auc',
        mode='max',
        save_best_only=True,
        verbose=1
    )
    
    # Entrenamiento con callbacks
    history = modelo.fit(
        X_train, y_train,
        batch_size=batch_size,
        epochs=epochs,
        validation_data=(X_val, y_val),
        callbacks=[early_stopping, reducir_lr, checkpoint],
        verbose=1
    )
    
    return modelo, history

def evaluar_modelo(modelo, X_test, y_test):
    """Evalúa el modelo y genera visualizaciones."""
    # Predicciones
    y_pred_prob = modelo.predict(X_test)
    y_pred = np.round(y_pred_prob)
    
    # Métricas
    print("\nInforme de clasificación:")
    print(classification_report(y_test, y_pred))
    
    conf_matrix = confusion_matrix(y_test, y_pred)
    print("\nMatriz de confusión:")
    print(conf_matrix)
    
    # Calcular curva ROC
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    
    # Visualizaciones
    plt.figure(figsize=(12, 5))
    
    # Gráfica de la matriz de confusión
    plt.subplot(1, 2, 1)
    plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion matrix')
    plt.colorbar()
    clases = ['No Apnea', 'Apnea']
    tick_marks = np.arange(len(clases))
    plt.xticks(tick_marks, clases, rotation=45)
    plt.yticks(tick_marks, clases)
    
    # Añadir valores a la matriz
    thresh = conf_matrix.max() / 2
    for i in range(conf_matrix.shape[0]):
        for j in range(conf_matrix.shape[1]):
            plt.text(j, i, format(conf_matrix[i, j], 'd'),
                     horizontalalignment="center",
                     color="white" if conf_matrix[i, j] > thresh else "black")
    
    plt.ylabel('Real')
    plt.xlabel('Predicted')
    
    # Gráfica de la curva ROC
    plt.subplot(1, 2, 2)
    plt.plot(fpr, tpr, color='k', linestyle='-', lw=2, label=f'ROC curve (area = {roc_auc:.2f})') # Separar color y estilo
    plt.plot([0, 1], [0, 1], color='k', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', color='black')
    plt.ylabel('True Positive Rate', color='black')
    plt.title('ROC Curve', color='black')
    plt.legend(loc="lower right", labelcolor='black')
    
    plt.tight_layout()
    plt.savefig('evaluacion_modelo_apnea_adaptativo.png')
    plt.show()
    
    return y_pred_prob, conf_matrix, roc_auc

def visualizar_predicciones(X_test, y_test, y_pred_prob, duraciones_test, indices=None):
    """Visualiza algunas predicciones para evaluar el rendimiento del modelo."""
    if indices is None:
        # Seleccionar aleatoriamente algunas ventanas para visualizar
        indices = np.random.choice(range(len(X_test)), min(5, len(X_test)), replace=False)
    
    plt.figure(figsize=(15, 10))
    for i, idx in enumerate(indices):
        plt.subplot(len(indices), 1, i+1)
        
        # Determinar la longitud real de la secuencia (excluyendo el padding)
        ventana = X_test[idx].flatten()
        # Identificar donde termina la señal real y empieza el padding (valores 0)
        longitud_real = duraciones_test[idx] if duraciones_test is not None else np.sum(ventana != 0)
        
        # Usar solo la señal real (sin padding)
        ventana_real = ventana[:longitud_real]
        tiempo = np.arange(len(ventana_real)) / FS_NUEVO  # Convertir a segundos
        
        plt.plot(tiempo, ventana_real, 'b-')
        
        # Colorear fondo según predicción
        pred_prob = y_pred_prob[idx][0]
        pred_class = 1 if pred_prob >= 0.5 else 0
        real_class = y_test[idx]
        
        color = 'green' if pred_class == real_class else 'red'
        label = f'Real: {"Apnea" if real_class else "Normal"}, '
        label += f'Pred: {"Apnea" if pred_class else "Normal"} ({pred_prob:.2f})'
        label += f', Duración: {longitud_real/FS_NUEVO:.1f}s'
        
        plt.title(label)
        plt.fill_between(tiempo, min(ventana_real), max(ventana_real), color=color, alpha=0.2)
        
        plt.xlabel('Tiempo (s)')
        plt.ylabel('Señal normalizada')
        plt.grid(True)
    
    plt.tight_layout()
    plt.savefig('visualizacion_predicciones_adaptativas.png')
    plt.show()

def main():
    # 1. Cargar datos
    print("Cargando dataset...")
    df = cargar_dataset()
    
    # 2. Crear ventanas adaptativas
    print("Creando ventanas adaptativas para el entrenamiento...")
    X, y, duraciones = crear_ventanas_adaptativas(df)
    
    # 3. Preparar datos para el modelo (aplicar padding)
    X_padded, y = preparar_datos_para_modelo(X, y)
    print(f"Forma de X después de padding: {X_padded.shape}")
    
    # 4. Dividir en conjuntos de entrenamiento, validación y prueba
    # Dividimos primero los índices para mantener la correspondencia con duraciones
    indices = np.arange(len(X_padded))
    indices_train_val, indices_test, y_train_val, y_test = train_test_split(
        indices, y, test_size=0.2, random_state=42, stratify=y
    )
    indices_train, indices_val, y_train, y_val = train_test_split(
        indices_train_val, y_train_val, test_size=0.2, random_state=42, stratify=y_train_val
    )
    
    # Ahora obtenemos los datos usando los índices
    X_train = X_padded[indices_train]
    X_val = X_padded[indices_val]
    X_test = X_padded[indices_test]
    
    # También mantenemos las duraciones originales para visualización
    duraciones_test = duraciones[indices_test]
    
    print(f"Conjunto de entrenamiento: {X_train.shape}")
    print(f"Conjunto de validación: {X_val.shape}")
    print(f"Conjunto de prueba: {X_test.shape}")
    
    # 5. Entrenar el modelo
    print("\nEntrenando modelo CNN-LSTM...")
    modelo, history = entrenar_modelo(X_train, y_train, X_val, y_val)
    
    # 6. Evaluar el modelo
    print("\nEvaluando modelo...")
    y_pred_prob, conf_matrix, roc_auc = evaluar_modelo(modelo, X_test, y_test)
    
    # 7. Visualizar algunas predicciones
    print("\nVisualizando predicciones...")
    visualizar_predicciones(X_test, y_test, y_pred_prob, duraciones_test)
    
    # 8. Guardar el modelo
    modelo.save('ignorar_cnn_lstm_apnea_adaptativo_fijo.h5')
    print("\nModelo guardado como 'modelo_cnn_lstm_apnea_adaptativo.h5'")
    
    # 9. Visualizar historia del entrenamiento
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training', linestyle='-', color='black')
    plt.plot(history.history['val_loss'], label='Validation', linestyle='--', color='black')
    plt.title('Loss during training', color='black')
    plt.xlabel('Epoch', color='black')
    plt.ylabel('Loss', color='black')
    plt.legend(labelcolor='black')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training', linestyle='-', color='black')
    plt.plot(history.history['val_accuracy'], label='Validation', linestyle='--', color='black')
    plt.title('Accuracy during training', color='black')
    plt.xlabel('Epoch', color='black')
    plt.ylabel('Accuracy', color='black')
    plt.legend(labelcolor='black')

if __name__ == "__main__":
    main()
```
